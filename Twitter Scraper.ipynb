```python
import pandas as pd
import twint
import nest_asyncio
nest_asyncio.apply()
import datetime
import time
import numpy as np
```

```python
def twitter(query = None, start = None, end = None, limit = None, user = None):
    config = twint.Config()
    config.Limit = limit
    config.Search = query
    config.Pandas = True
    config.Hide_output = True
    config.Since = start
    config.Until = end
    config.Username = user
    twint.run.Search(config)
    df = twint.output.panda.Tweets_df
    df['date'] = pd.to_datetime(df['date'])
    return df
```

```python
#example use case
SABIC = twitter(user = 'SABIC', limit = 10)
```
